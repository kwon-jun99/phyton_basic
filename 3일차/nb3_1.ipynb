```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))

print(f[1])

f[1]= "abc"
print(f)
```

```python
# 7. 튜플 (tuple)
g = (1, 2, 3)
print(g)
print(type(g))

print(g[1])

#g[1]=9
```

```python
# 8. 딕셔너리 (dict)
h = {"id": "admin", "pw": "1234"}
print(h)
print(type(h))

print(h["id"])

h["pw"]="0000"
print(h)
```

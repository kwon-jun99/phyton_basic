```python
import random #random.random() 함수만으로

#로또
# 1. 공을 45개 만든다 -> 변수 45개
balls= list(range(1,46))
print(balls)

# 2. balls -> shuffle
for i in range(0,1000000):
    temp=balls[0]
    r_num=int(random.random() * len(balls)) #0.000~44.99999... 
    #r_num=math.floor(random.random() * len(balls))
    balls[0]=balls[r_num]
    balls[r_num]=temp

print(balls)
print(balls[:6])
```

```python
def my_lotto_create(a):
    result=[]
    for i in range(a):
        lotto=random.sample(range(1,46),6)
        lotto.sort()
        #print(f"{i+1}번째 로또번호: {lotto}")
        result.append(tuple(lotto))
    return tuple(result)

result= my_lotto_create(10)
print(result)
```

```python
def dice_cast():                #주사위 던지는 행위
    d1= random.randint(1,6)
    d2= random.randint(1,6)
    return d1+d2

person1_dice = dice_cast()
person2_dice = dice_cast()

if person1_dice == person2_dice:
    print(f"비겼다. person1:{person1_dice},person2:{person2_dice}")
elif person1_dice > person2_dice:
    print(f"person1_dice Win. person1:{person1_dice},person2:{person2_dice}")
else:
    print(f"person2_dice Win. person1:{person1_dice},person2:{person2_dice}")
```

```python
import datetime
import time
```

```python
now= datetime.datetime.now()
print(now)
print(now.year)
print(now.month)
print(now.day)
```

```python
today= datetime.date.today()
print(today)
```

```python
# print(today.weekday()) #오늘날짜 출력

weekday=today.weekday()

week=["월","화","수","목","금","토","일"] #0="월"

print(week[weekday])

xmas= datetime.date(2026,12,25)
print(week[xmas.weekday()])
```

```python
print(xmas-today)

print(today + datetime.timedelta(days=1000))      # print(today + 1000) = X 1000일인지 1000달인지 알수없음
```

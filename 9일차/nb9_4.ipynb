```python
def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

```python
def square(x):
    return x * x

def test(fun):      
    return fun(5)

print(test(square))
```

```python
def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()
outer()()
```

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```

```python
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
names = ["홍길동", "이순신", "강감찬"]

for i, name in enumerate(names):
    #print(i, name)  
    print(f"({i+1}) {name}")
```

```python
nums = [0, 0, 3]
print(any(nums))

nums = [1, 2, 3]
print(all(nums))

```

```python
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
scores2 = [100,100,100]
result = zip(names, scores, scores2)
print(list(result)) # [('홍길동', 90), ('이순신', 80), ('김유신', 100)]
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    # 한 줄씩 읽기.
    for line in f: 
        print(line) # 라인 끝의 줄바꿈이 문자 포함되어 두번 줄바꿈이 일어난다(print함수도 줄바꿈이 기본값)
        print(type(line))
```

```python
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장
names=[]
scores=[]
with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
       # print(name, int(score))
        names.append(name)
        scores.append(score)

print(names)
print(scores)
```

```python
students = {}

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    #reader(): csv모듈 안에 있는 함수
    reader = csv.reader(f)
    #print(type(reader)) <class '_csv.reader'>

    for row in reader:
        print(row)
        #print(type(row)) <class 'list'>
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인은 키, 데이터는 키의 값으로

    for row in reader:
        print(row)
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)
data[2]["영어"]=80
print(data)
```

```python
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  
```

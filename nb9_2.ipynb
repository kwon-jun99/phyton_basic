```python
# 전역 변수 : 함수 정의와 같은 코드 레벨에 있는 변수
x = 10 

def fun_global():
    print(x) # x 전역변수를 사용

def fun_local():
    x = 333
    print(x) # 지역변수 x가 존재하기 때문에 지역변수 x를 사용


fun_global()
fun_local()
```

```python
def my_intersect(a,b)->list:        #->list= 반환타입을 list로 고정
    '''
    a:리스트
    b:리스트
    반환타입: 리스트
    내용: 두개의 리스트의 중복되는 값만을 새로운 리스트로 반환
    '''
    result=[]
    for e in a:
        if e in b:
            result.append(e)
    return result

a=[1,2,3,4,5,6,7,8]
b=[7,8,9]
r1=my_intersect(a,b)
print(r1)
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 공통 요소 개수

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# a에는 있고 b에는 없는 요소

def my_difference(a, b):
    result = []
    for e in b:
        if e not in a:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result:
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```

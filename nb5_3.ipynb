```python
#인덱싱

nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

n2= (3,5)
print(n2[0])
# n2[0]=999 튜플은 불변객체 변경불가
```

```python
#슬라이싱

nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50]

# 2칸씩 건너뛰기

nums=(10,20,30,40,50,60,70,80)
print(nums[::2])

# 2칸씩 건너뛰기(2번째부터)

nums=(10,20,30,40,50,60,70,80)
print(nums[1::2])

#역순 출력

nums=[10,20,30,40,50]
print(nums[::-1])
```

```python
a = [10, 20, 30]
b = a #공동참조이다.

print(a)
print(b)
print(a == b)
print(a is b)

b[0] = 999
print(a) 
print(b) 

a = [10, 20, 30]
b = a[:] #공동참조 아니다.

print(a)
print(b) 
print(a == b) 
print(a is b)

b[0] = 999
print(a) # [10, 20, 30]
print(b) # [999, 20, 30]
```

```python
nums = [10, 20, 30, 40]

print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
```

```python
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)

# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```

```python
#튜플

t = 1, 3, 5 # 팩킹 실행 (1,3,5)
print(t) 

print(type(t))

x, y, z = t # 언팩킹 실행
print(x, y, z)
print(type(t))

#리스트도 팩킹, 언팩킹 가능하다. 단, 튜플로 변경된 후 작업진행됨.

#패킹, 언패킹시 *연산자
#a,b=[1,2,3,4] #too many values to unpack
*a,b=[1,2,3,4]
print(a)
print(b)

c,*d=(1,2,3,4,)
print(c)
print(d)

e,*f,g=(1,2,3,4)
print(e)
print(f)
print(g)

h=[1,2,3]
print(h)
print(*h) #print(h[0],h[1],h[2])
```

```python
# 셋(set): 중복을 허용하지 않는 자료형,순서가 없다(인덱싱 불가)

a={1,2,3,4}
print(a)
# print((a)[0]) # 셋은 인덱스값이 없다. 순서가 없다.
print(list(a)[0])
```

```python
a={"1":  "admin", "pw": "1234"}
print(a,type(a))

d={} #{}은 빈 set이 아닌 빈 dict 자료형이다. 빈셋이 필요하면 set() 함수 사용.
print(d)
print(type(d))

s=set()
print(s)
print(type(s))
```

```python
#셋의 집합연산

a={1,2,3}
b={3,4,5}

#교집합 
print(a&b)
print(a.intersection(b))

#합집합
print(a|b)
print(a.union(b))

#차집합
print(a-b)
print(a.difference(b))
```

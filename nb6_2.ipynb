```python
# 2단계: 데이터 전처리단계

# 1.데이터 갯수확인 

len(students)
```

```python
# 2단계: 데이터 전처리단계

# 2.결측지 확인

missing_count = 0 # 결측지가 몇개인지?

for student in students:
    if student["성적"] is None:
        missing_count += 1

print("결측치 개수:", missing_count)
```

```python
# 2단계: 데이터 전처리단계

# 3.결측지 처리(평균값 대체: 어떤 값으로 대처하는지는 요구사항에 따라 다름

scores = []

for student in students:
    if student["성적"] is not None:
        scores.append(student["성적"])

avg = sum(scores) / len(scores)

for student in students:
    if student["성적"] is None:
        student["성적"] = avg

print(students)
```

```python
# 2단계: 데이터 전처리단계

# 4.이상치 탐지(정상 범위: 0~100점)

for student in students:
    score = student["성적"]

    if score < 0 or score > 100:
        print(student)
```

```python
# 2단계: 데이터 전처리단계

# 5.이상치 처리(어떻게 처리할지는 요구사항에 따라 다름)

for student in students:

    if student["성적"] > 100:
        student["성적"] = 100

    elif student["성적"] < 0:
        student["성적"] = 0

students
```

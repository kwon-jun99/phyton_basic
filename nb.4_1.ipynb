```python
# 1) 문자 채우기, 정렬
name = "Python"

print(f"{name:*<15}")  # 왼쪽 정렬, 빈칸은 *로 채움
print(f"{name:*^15}")  # 가운데 정렬, 빈칸은 *로 채움
print(f"{name:*>15}")  # 오른쪽 정렬, 빈칸은 *로 채움
# 변수 사용 가능
w = 15
print(f"{name:*>{w}}")  # 오른쪽 정렬, 빈칸은 *로 채움
```

```python
name = "Park"
age = 30

# f-string 기본 사용
print(f"이름: {name}, 나이: {age}")  # 변수 값을 문자열 안에 직접 삽입

# 소수점 자리수 지정
pi = 3.141592
print(f"{pi:.2f}")  # 소수점 둘째 자리까지 출력 → 3.14

# 자리수(폭) 지정
num = 7
print(f"{num:5}")   # 전체 5칸 확보, 오른쪽 정렬 → '    7'
print(f"{num:05}")  # 빈칸을 0으로 채움 → 00007
```

```python
# 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"{'Python':!<10}")  # 왼쪽 정렬
print(f"{'Python':!>10}")  # 오른쪽 정렬
print(f"{'Python':!^10}")  # 가운데 정렬
```

```python
# 기본 사용
print("이름: {}, 나이: {}".format("Park", 30))  # {} 순서대로 값 들어감

# 인덱스 사용
print("이름: {0}, 나이: {1}".format("Park", 30))  # 번호로 위치 지정

# 이름 지정
print("이름: {name}, 나이: {age}".format(name="Park", age=30))  
# 변수 이름으로 직접 지정 가능
```

```python
name = "Park"
age = 30

print("이름: %s, 나이: %d" % (name, age))  
# %s = 문자열, %d = 정수

pi = 3.141592
print("%.2f" % pi)  # 소수점 둘째 자리까지 출력 → 3.14
```

```python
# 2) 숫자 채우기
num = 123

print(f"{num:08}")   # 전체 8자리 문자로 변경 후, 부족한 자리는 0으로 채움
print(f"{num:010}")
```

```python
# 3) 부호 포함 출력
a = 150
b = -50

print(f"{a:+}")  # 문자열 변경시 양수는 +, 음수는 - 표시
print(f"{b:+}")    # 음수는 - 표시
```

```python
# 5) 천 단위 구분기호 출력
sales = 1234567890
print(f"{sales:,}")  # 천 단위마다 콤마 추가
```
